## Exploratory data analysis

```python
store.groupby('Sub-Category')[['Profit','Sales']].agg(['sum']).plot.bar()
plt.show()
```

- From this graph we can say that Tables subcategory sales are more but profit is neagtive when it compared with paper the sales arevery low but the profit is postive

```python
store.groupby('Sub-Category')[['Profit','Quantity']].agg(['sum']).plot.bar()
plt.title('Quantity and profit relation based on sub category')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(store['Ship Mode'].unique(),store.groupby(['Ship Mode'])['Sales'].sum())
plt.title('shipmode wise sales')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.bar(store['Ship Mode'].unique(),store.groupby(['Ship Mode'])['Profit'].sum())
plt.title('shipmode wise profit')
plt.xticks(rotation=45)
plt.show()

        
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(x=store['Segment'].unique(),height=store.groupby(['Segment'])['Sales'].sum(),color=['purple','orange','green'])
plt.title('Segment wise sales')
plt.xticks(rotation=45)

plt.subplot(1,3,2)
plt.bar(x=store['Segment'].unique(),height=store.groupby(['Segment'])['Profit'].sum(),color=['purple','orange','green'])
plt.title('Segment wise profit')
plt.xticks(rotation=45)
plt.show()

plt.subplot(1,3,3)
plt.bar(x=store['Segment'].unique(),height=store.groupby(['Segment'])['Discount'].sum(),color=['purple','orange','green'])
plt.title('Segment wise Discount')
plt.xticks(rotation=45)
plt.show()
```

- From the graph we can say that consumer segment has high sales and profits
- Discount for consumer also high so the profit and sales hit maximum
- so we have to provide discount for corporate and homeoffice to reach max sales and profit


```python
# total sales and profit in region
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.pie(store.groupby(['Region'])['Sales'].sum(),labels=['Central','East','South','West'],autopct='%1.2f%%',shadow=True)
plt.title('Region wise sales')


plt.subplot(1,2,2)
plt.pie(store.groupby(['Region'])['Profit'].sum(),labels=['Central','East','South','West'],autopct='%1.2f%%',shadow=True)
plt.title('Region wise profit')

plt.show()
```

from above graph we can conclude that Central region is generating least profit

```python
# total sales and profit in Category
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.pie(store.groupby(['Category'])['Sales'].sum(),explode=[0,0.2,0],labels=['Furniture', 'Office Supplies', 'Technology'],autopct='%1.2f%%',shadow=True)
plt.title('Category wise sales')


plt.subplot(1,2,2)
plt.pie(store.groupby(['Category'])['Profit'].sum(),explode=[0.2,0,0],labels=['Furniture', 'Office Supplies', 'Technology'],autopct='%1.2f%%',shadow=True)
plt.title('Category wise profit')

plt.show()
```

```python
# no of orders  by subcategory
plt.figure(figsize=(20,10))
sns.countplot(x=store['Sub-Category'],order=store['Sub-Category'].value_counts().index,color='skyblue')
plt.show
```

- from binders subcategory most of the orders were placed followed by papers and furnishings.



```python
# no of orders by city
plt.figure(figsize=(15,5))
sns.countplot(x=store['City'],order=store['City'].value_counts().head(20).index,color='skyblue')
plt.xticks(rotation=45)
plt.show
```

```python
# no of orders by state
plt.figure(figsize=(15,10))
sns.countplot(x=store['State'],order=store['State'].value_counts().head(20).index,color='purple')
plt.xticks(rotation=45)
plt.show
```

-  Most of the orders from newyork city and state

```python
sns.lineplot(x='Discount',y='Profit',data=store)
plt.show()
```
